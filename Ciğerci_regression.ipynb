```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/fifa_players.csv")
```

```python
df.head()
```

```python
yc = df[["height_cm", "weight_kgs", "heading_accuracy"]]
```

```python
yc
```

```python
X = yc.drop(columns= ["heading_accuracy"],axis=1)
y = yc[["heading_accuracy"]]
```

```python
import statsmodels.api as sm

lm = sm.OLS(y, X)
model = lm.fit()
model.summary()
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
model =lm.fit(X, y)
print(f"B0: {model.intercept_}")
print(f"B1,2,3 values: {model.coef_}")
```
